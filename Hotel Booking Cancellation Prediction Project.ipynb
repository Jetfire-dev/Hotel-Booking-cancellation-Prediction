```python
#importing neccessary modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#importing dataset
df= pd.read_csv(r"C:\Users\IIJet\Downloads\booking.csv")
```

```python
df
```

```python
#checking null data 
df.isnull().sum()
```

```python
# using replace function to perform label encoding 
df["booking status"]= df["booking status"].replace("Canceled",1)
df["booking status"]= df["booking status"].replace("Not_Canceled",0)
```

```python
df
```

```python
# Checking summary of data
df.info()
```

```python
# Dropping column booking ID from dataframe
df.drop(columns=['Booking_ID'], inplace=True)
```

```python
# Checking specific date 2018-2-29 because based on research 2018 is not a leap year until 2020
df[df["date of reservation"]=="2018-2-29"]
```

```python
# Since now that there is false date in the data, it is best to get review of date of reservation column by checking through years.
# This will give good idea of what to expect from date/time format and to see if theres any reason to include in features

df= df[df['date of reservation'] != '2018-2-29']

df['date of reservation'] = pd.to_datetime(df['date of reservation'])

df.groupby([df['date of reservation'].dt.year,df['date of reservation'].dt.month])['date of reservation'].count()
```

```python
# since date of reservation has little role to play in terms of influencing booking cancellation
# also to understand the above grouping of y m d format, we can see 2015 and 2016 with little booking 
# while 2017-2018 has had more booking throughout the year including false date in which 2018 is not a leap year
# in terms of relevancy, date of reservation will be removed for the purpose of sensitive data when needed for predictive modeling

df.drop(columns=['date of reservation'], inplace=True)
df
```

```python
# Using get dummies for one hot encoding to seperate categorical values into columns with binary data

encoded_df = pd.get_dummies(df, columns=['market segment type', 'room type', 'type of meal'], prefix= ["Market type","rooms","meals"])

encoded_df
```

```python
# After data pre processing and manipulation, we check the summary of cleaned data

encoded_df.info()
```

```python
# Feature variable includes everything except booking status column

x= encoded_df.drop("booking status",axis=1)
```

```python
# target variable is booking status for cancellation and non cancellation

y= encoded_df["booking status"]
```

```python
#import train test split from sci-kit library
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# checking the size of train and test set
print("Train data shape of x: ", x_train.shape, "y: ",y_train.shape)
print("Test data shape of x: ", x_test.shape, "y: ",y_test.shape)
```

```python
# Using correlation heatmap for data visualisation to display encoded data before moving to machine learning algorithm models

df1 = pd.concat ([x,y],axis=1)

corr_matrix = df1.corr()

plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Based on target variable, Classification models will be the choice here due to binary values here
# Not only that, since I am predicting cancellation booking, I will need to find out which classification algorithm is accurate
# The higher the accuracy the better the model for this dataset, any other model would be unfit if they dont have high accuracy
# Starting with support vector machine model which is used for both regression and classification

from sklearn.svm import SVC

svc=SVC()
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
y_pred
```

```python
# Now that model is built and fit into prediction, next is Model evaluation which will help in learning the accuracy of the model
# in order to get simple idea of both train and test data, we will start with bias and variance

bias=svc.score(x_train,y_train)
bias
```

```python
variance= svc.score(x_test,y_test)
variance
```

```python
# Although they have almost same score, its still best to compare with other evaluation metrics including cross validation

from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
acc
```

```python
#import kfold for cross validation and cross val score for cross validation results
# Since the dataset is very large kfold would be more preferrable as more splits can be done for better accuracy
from sklearn.model_selection import cross_val_score, KFold

k=15

kfold = KFold(n_splits=k, shuffle=True, random_state=42)
```

```python
# Now to find the result of cross validation
cv_results = cross_val_score(svc, x, y, cv=kfold, scoring='accuracy')

print("Cross validation results: ", cv_results)
print("Mean accuracy: ", cv_results.mean())
print("Standard deviation of accuracy: ", cv_results.std())
```

```python
#importing confusion matrix for predicted and actual labels
from sklearn.metrics import confusion_matrix

cm= confusion_matrix(y_test,y_pred)
cm
```

```python
# using heatmap for visualising confusion matrix of support vector classifier for true and false positives and negatives
sns.heatmap(cm, annot=True, fmt='d')
plt.title("Confusion matrix for Support vector Classifier")
plt.xlabel("Prediction Label")
plt.ylabel("Actual Label")
plt.show()
```

```python
#Classification report for support vector classifier model
from sklearn.metrics import classification_report
rt=classification_report(y_test,y_pred)
print("Classification report:\n", rt)
```

Now that The SVC model is done with some changes such as splits and comparing cross validation technique with other evaluation metrics and with that above i can conclude SVC model is 77% accurate with prediction for booking cancellation. 

Conclusion: The model is decently accurate with 77% however lets continue with other classification algorithms for better comparison of results.
